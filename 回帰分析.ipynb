```python
from sklearn import datasets
```

```python
boston.head()
```

```python
x = boston.values
```

```python
x.shape
```

```python
y = boston_data.target
```

```python
y.shape
```

```python
clf = linear_model.LinearRegression(normalize=True)
```

```python
clf.fit(x, y)
```

```python
clf.coef_.shape
```

```python
pd.DataFrame({
    "name": boston.columns,
    "coef": clf.coef_
}).sort_values(by='coef')
```
